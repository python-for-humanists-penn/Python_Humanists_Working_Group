```python
import pandas
import umap
import textwrap

import matplotlib
%matplotlib inline
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource, CategoricalColorMapper
from bokeh.palettes import magma

output_notebook()
```

```python
topic_data_path = "/Users/jdummer/Desktop/royalsociety_corpus/plaintext/plaintext_output/output_csv/topics-metadata.csv"
```

```python
topic_data = pandas.read_csv(topic_data_path)
```

```python
topic_data.iloc[0:3, 0:10]
```

```python
metadata = topic_data.iloc[:, 0:6]
topics = topic_data.iloc[:, 6:]
```

```python
umap_2d = umap.UMAP()
vis_2d = umap_2d.fit_transform(topics.values)
vis_2d
```

```python
vis_2d.shape
```

```python
magma_palette = magma(6)[::-1]
sp = pandas.DataFrame({'umap_1': vis_2d[:, 0],
                       'umap_2': vis_2d[:, 1],
                       'color': [magma_palette[int(t * 6)] 
                       for t in topic_data['8 fish birds species']],
                       'title': ['<br>'.join(textwrap.wrap(t)) for t in metadata['title']]})
sp.iloc[0:10]
```

```python
plot_figure = figure(
    title='UMAP projection the Royal Society Corpus (18th century)',
    plot_width=600,
    plot_height=600,
    tools=('pan, wheel_zoom, reset')
)

plot_figure.add_tools(HoverTool(tooltips="""

    <span style='font-size: 8px; color: #224499'>Title:</span>
    <span style='font-size: 8px'>@title{safe}</span>
"""))

plot_figure.circle(
    'umap_1',
    'umap_2',
    color='color',
    source=sp,
)

show(plot_figure) 
```
